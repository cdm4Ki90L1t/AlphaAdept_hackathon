**Обучение модели на данных о свечах акций**

Данная функция используется для обучения модели на данных о свечах акций. Она принимает на вход DataFrame с данными о свечах и целочисленное значение i, которое используется для определения цены закрытия через i дней в будущем.

```python
 def train_model(df, i,zn):
    #Отделяем data frame для последних 30 дней, которые не будут задействован в обучении
    df_copy = df.copy()
    subset = df_copy.iloc[len(df_copy) - 30:len(df_copy) + 1]
    df_copy = df_copy.drop(subset.index)

    df_copy['price'] = df_copy['close'].shift(-1 * i - 1)
    if zn == 1:
        df_copy['Target'] = (df_copy['close'].shift(-1 * i - 1) > df_copy['close']).astype(int)
    else:
        df_copy['Target'] = (df_copy['close'].shift(-1 * i - 1) < df_copy['close']).astype(int)
    # Удаляем строки с пропущенными значениями
    drop = df_copy.dropna()

    # Выбираем признаки для модели
    if indicator == 'stochrsi':
        features = drop[['close', 'open', 'high', 'low', 'volume', 'STOCHRSIk_14_14_3_3', 'STOCHRSId_14_14_3_3']]
    elif indicator == 'aroon':
        features = drop[['close', 'open', 'high', 'low', 'volume', 'AROOND_14', 'AROONU_14', 'AROONOSC_14']]
    else:
        features = drop[['close', 'open', 'high', 'low', 'volume', indicator]]

    # Разделяем данные на обучающий и тестовый наборы
    X_train, X_test, y_train, y_test = train_test_split(features, drop['price'], test_size=0.2, random_state=42)
    X_train1, X_test1, y_train1, y_test1 = train_test_split(features, drop['Target'], test_size=0.2, random_state=42)

    # Создаем и обучаем модель линейной регрессии
    model = LinearRegression()
    model.fit(X_train, y_train)

    logistic_model = LogisticRegression()
    logistic_model.fit(X_train1, y_train1)

    # Предсказываем вероятность успешного прогноза для i Дней
    predictions = model.predict(X_test)
    prob_predictions = logistic_model.predict_proba(X_test1)[:, 1]
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    # Получаем значение насколько предсказание верно
    accuracy = accuracy_score(y_test1, (prob_predictions > 0.5).astype(int))

    return accuracy, rmse, model, logistic_model
```

Функция создает копию исходного DataFrame, выбирает последние 30 строк (последние 30 дней) и удаляет их из копии. Затем функция создает новые столбцы в DataFrame: 'price' - цена закрытия через i дней в будущем, и 'Target' - бинарный признак, который равен 1, если цена закрытия через i дней в будущем выше, чем цена закрытия текущего дня, и 0 в противном случае.

Далее функция выбирает только те столбцы из DataFrame, которые используются в качестве признаков для обучения модели.

Затем функция разбивает данные на обучающую и тестовую выборки с помощью функции train_test_split из библиотеки sklearn. Для каждой выборки создаются отдельные DataFrame для признаков и целевой переменной.

Позже функция создает две модели: линейную регрессию и логистическую регрессию. Модели обучаются на обучающих данных с помощью метода fit.

Наконец, функция делает предсказания на тестовых данных с помощью метода predict для линейной регрессии и метода predict_proba для логистической регрессии. Для логистической регрессии используется вероятность принадлежности к классу 1 (цена закрытия через i дней в будущем выше, чем цена закрытия текущего дня).

О коде:


*   Класс Backtest используется для проведения backtest'а на данных data с использованием стратегии из класса SmaCross. Параметры cash и commission определяют начальный капитал и комиссию за сделку соответственно. Параметр exclusive_orders установлен в True, что означает, что только одна операция может быть открыта в одно и то же время.
*   Метод run() запускает backtest и возвращает объект stats, содержащий результаты тестирования.
*   Метод _trades объекта stats возвращает список сделок, совершенных во время backtest'а.
*   Метод plot() объекта bt используется для построения графика результатов backtest'а.












```python
import backtesting as bt
from backtesting import Strategy, Backtest
from backtesting.lib import crossover
import pandas as pd

def MovingAverage(closes:pd.Series, n:int) -> pd.Series:
    return pd.Series(closes).rolling(n).mean()
```

```python
class SmaCross(Strategy):
    sma_fast = 12
    sma_slow = 24

    def init(self):
        self.sma1 = self.I(MovingAverage, self.data.Close, self.sma_fast)
        self.sma2 = self.I(MovingAverage, self.data.Close, self.sma_slow)

    def next(self):
        if self.data['signal_buy']==True:
            self.buy()
        elif self.data['signal_sell']==True :
            self.position.close()
```

Входные параметры:


*   индикатор - rsi
*   акция - SBER
*   длительность работы стратегии - 360 дней

```python
data = df_merged_all
bt = Backtest(data, SmaCross, cash=10_000, commission=0,exclusive_orders=True)
stats = bt.run()
print(stats)
print(stats._trades)
bt.plot()
```

```python
def stat(df,i):
    #смешаем data frame на i дней, чтобы собрать данные будет ли цена расти через i Дней и для предсказания цены через i дней
    df['price'] = df['close'].shift(-1*i-1)
    df['Target'] = (df['close'].shift(-1*i-1) > df['close']).astype(int)

    # Удаляем строки с пропущенными значениями
    drop = df.dropna()

    # Выбираем признаки для модели
    features = drop[['open', 'STOCHRSIk_14_14_3_3','STOCHRSId_14_14_3_3', 'AROOND_14' ,'AROONU_14','AROONOSC_14']]

    # Разделяем данные на обучающий и тестовый наборы
    X_train, X_test, y_train, y_test = train_test_split(features, drop['price'], test_size=0.2, random_state=42)
    X_train1, X_test1, y_train1, y_test1 = train_test_split(features, drop['Target'], test_size=0.2, random_state=42)

    # Создаем и обучаем модель линейной регрессии
    model = LinearRegression()
    model.fit(X_train, y_train)
    logistic_model = LogisticRegression()
    logistic_model.fit(X_train1, y_train1)

    # Предсказываем процентное изменение цены для тестового набора
    predictions = model.predict(X_test)
    prob_predictions = logistic_model.predict_proba(X_test1)[:, 1]

    # Оцениваем точность модели для цены
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)

    # Получаем значение насколько предсказание верно
    accuracy = accuracy_score(y_test1, (prob_predictions > 0.5).astype(int))

    return accuracy,rmse, model, logistic_model
```

```python
for j in range(per):
    #==================_ЧАСТЬ_1_==================
    # берем сабсет за последний год и в дальнейшем будем уменьшать размер сабсета
    current_subset = df.iloc[len(df) - per + j : len(df) + 1]
    # удаляем из исходных данных то, что перешло в сабсет
    dff = df.drop(current_subset.index)
    # берем только первое значение, и в дальнейшем это число будет прибавлять 1 день
    current_subset = current_subset.iloc[0:1]
    current_subset2 = df.iloc[len(df) - per + j - 30 : len(df) - per + j]

    # Тренируем модель для предсказывания цен до 30 дней
    for i in range(30):
        accuracy, rmse, _, _ = stat(dff, i)
        data[str(i)] = [accuracy, rmse, i + 1]

    # Получаем наиболее вероятный успешный исход для дней от 0 до 30
    max_accuracy_key = max(data, key=lambda k: data[k][0])

    # Используя модели делаем прогнозы на количество дней указанных в data[max_accuracy_key][2]
    _, _, model, logistic_model = stat(dff, data[max_accuracy_key][2])

    # стоит отметить, что сабсет для тренировки модели имеет только Open
    # это сделано для того, чтобы программа только воспренимала данные как будто она берет данные в начале дня
    # в ином случае получается то, что мы в начале дня получаем все данные за конец дня сразу, а это логическая ошибка

    # прогоняем данные через модель
    # также отмечу, что тут не берется 1 индикатор, а используются все сразу
    predictions = model.predict(
        current_subset2[
            [
                "open",
                "STOCHRSIk_14_14_3_3",
                "STOCHRSId_14_14_3_3",
                "AROOND_14",
                "AROONU_14",
                "AROONOSC_14",
            ]
        ]
    )
    prob_predictions = logistic_model.predict_proba(
        current_subset[
            [
                "open",
                "STOCHRSIk_14_14_3_3",
                "STOCHRSId_14_14_3_3",
                "AROOND_14",
                "AROONU_14",
                "AROONOSC_14",
            ]
        ]
    )

    #==================_ЧАСТЬ_2_==================
    '''это параметр нужен, чтобы отслеживать, когда продавать открытую позицию
    либо при достижении срока сигнала, либо при достижении цены сигнала'''
    if bay == 1:
        srok -= 1
        if bay == 1 and current_subset["high"].iloc[-1] >= price:
            balance += price * 1000 - (price * 1000)*0.005
            bay = 0
            mas_buy.append((str(current_subset.index[0])[:10]))
        elif srok <= 0:
            balance += current_subset["open"].iloc[-1] * 1000 - (current_subset["open"].iloc[-1] * 1000)*0.005
            bay = 0
            mas_buy.append((str(current_subset.index[0])[:10]))
        continue

    '''этот параметр активируется, если выполнены параметры по предсказанной вероятности
    и тому, что предсказанная цена будет выше, чем текущая'''
    procent = (predictions[-1] - current_subset["open"].iloc[-1])/ current_subset["open"].iloc[-1]
    if prob_predictions[0][1] > 0.5 and current_subset["open"].iloc[-1] < predictions[-1] and bay == 0 and procent > 0.02:
        bay = 1
        balance -= current_subset["open"].iloc[0] * 1000 - (current_subset["open"].iloc[0] * 1000)*0.005
        srok = data[max_accuracy_key][2]
        price = predictions[-1]
        mas_sell.append((str(current_subset.index[0])[:10]))

# Выводим результат
dfff = df.iloc[len(df) - per : len(df) + 1]
print(balance)
```

Проводим бектест и смотрим результаты

```python
import backtesting as bt
from backtesting import Strategy, Backtest
from backtesting.lib import crossover
import pandas as pd

def MovingAverage(closes:pd.Series, n:int) -> pd.Series:
    return pd.Series(closes).rolling(n).mean()
```

```python
class SmaCross(Strategy):
    sma_fast = 12
    sma_slow = 24

    def init(self):
        self.sma1 = self.I(MovingAverage, self.data.Close, self.sma_fast)
        self.sma2 = self.I(MovingAverage, self.data.Close, self.sma_slow)

    def next(self):
        if self.data['signal_buy']==True:
            self.buy()
        elif self.data['signal_sell']==True :
            self.position.close()
```

========================**Стратегия №3**========================

```python
data = df_merged_all
bt = Backtest(data, SmaCross, cash=10_000, commission=0,exclusive_orders=True)
stats = bt.run()
print(stats)
print(stats._trades)
bt.plot()
```
